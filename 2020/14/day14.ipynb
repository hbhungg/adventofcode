```python
from numpy import binary_repr
with open('input.txt') as f:
    lines = f.read().split('\n')
```

```python
import copy
from itertools import product
def writemem(val, mask):
    retval = []
    for i, x in enumerate(mask):
        if x == '0':
            retval.append(val[i])
        else:
            retval.append(x)
    #print(val)
    #print(mask)
    #print(''.join(retval))
    retval = floating(retval)
    #print(retval)
    #print()
    return retval

def floating(val):
    retval = []
    r = val.count('X')
    ll = list(product([0,1], repeat=int(r)))
    for i in ll:
        temp = val.copy()
        for j in i:
            for i, k in enumerate(temp):
                if k == 'X':
                    temp[i] = str(j)
                    break
        ttemp = ''.join(temp)
        retval.append(int(ttemp, 2))
    return retval
```

```python
mem = {}
for i in lines:
    act, _, val = i.split()
    if act == 'mask':
        mask = val
    else:
        add = int(act[4:-1])
        v = binary_repr(add, 36)
        retval = writemem(v, mask)
        for i in retval:
            mem[str(i)] = int(val)
```
