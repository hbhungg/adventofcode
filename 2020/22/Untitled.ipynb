```python
with open('cards.txt') as f:
    lines = f.read().split()
```

```python
for p, i in enumerate(lines):
    if i == 'Player' or i == '1:' or i == '2:':
        lines.pop(p)

for p, i in enumerate(lines):
    if i == 'Player' or i == '1:' or i == '2:':
        lines.pop(p)
```

```python
p1 = [int(p) for p in lines[0:25]]
p2 = [int(p) for p in lines[25:]]
```

```python
print(p1)
print(p2)
```

```python
class Player:
    def __init__(self, name, startingHand):
        self.cards = startingHand
        self.name = name
    
    def __str__(self):
        return self.name +'`s Deck: ' + str(self.cards)
    
    def play_card(self):
        return self.cards.pop(0)
    
    def take_card(self, winCard, loseCard):
        self.cards.append(winCard)
        self.cards.append(loseCard)
    
    def status(self):
        if len(self.cards) == 0:
            return False
        return True
```

```python
a1 = Player("Player1", p1)
a2 = Player("Player2", p2)
def game():
    while a1.status() and a2.status():
        print(a1)
        print(a2)
        p1c = a1.play_card()
        p2c = a2.play_card()
        print(a1.name, 'plays: ', p1c)
        print(a2.name, 'plays: ', p2c)
        if p1c > p2c:
            print(a1.name, "wins")
            a1.take_card(p1c, p2c)
        elif p2c > p1c:
            print(a2.name, "wins")
            a2.take_card(p2c, p1c)
```

```python
game()
```
