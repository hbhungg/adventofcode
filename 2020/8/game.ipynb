```python
with open('input.txt') as f:
    l = f.read().split('\n')
```

```python
ls = [[x[0:3], x[3:], 0] for x in l]
for i, r in enumerate(ls):
    print(i, r)
```

```python
for i, e in enumerate(ls):
    if int(e[1]) + i > 621 and e[0] != 'acc':
        print(i, e, int(e[1])+i)
```

```python
def reset(ins):
    for i in ins:
        i[2] = 0

def run(ins):
    accum = 0
    head = 0
    try:
        while ins[head][2] < 2 and head < len(ins):
            value = int(ins[head][1])
            ins[head][2] += 1
            if ins[head][0] == 'acc':
                accum += value
                head += 1
            elif ins[head][0] == 'jmp':
                #print(head, ins[head], head+value)
                head += value
            elif ins[head][0] == 'nop':
                #print(head, ins[head], head+1)
                head += 1
    except IndexError as e:
        return accum, True
    return accum
```

```python
reset(ls)
print(run(ls))
```

```python
import copy

reset(ls)
for i in range(len(ls)):
    if ls[i][0] == 'jmp':
        temp = copy.deepcopy(ls)
        temp[i][0] = 'nop'
        print(run(temp))
    elif ls[i][0] == 'nop':
        temp = copy.deepcopy(ls)
        temp[i][0] = 'jmp'
        print(run(temp))
```
